```python
import pylab as plt
z = np.random.randint(255, size=(255,255))
print(np.mean(z))
plt.imshow(z, cmap='gray', interpolation='nearest')
plt.show()
```

```python
import pylab as plt
z = np.random.randint(255, size=(255,255))
y = np.mean(z)
z[z>(y/9)]=0
plt.imshow(z, cmap='gray', interpolation='nearest')
plt.show()
```
