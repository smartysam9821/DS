```python
fig = plt.figure(figsize=(10,7))
ax = plt.bar(gender_count.index, gender_count.PassengerId)
ax[0].set_color('green')

for num, count in enumerate(gender_count.PassengerId):
    plt.text(x=num, y=count+10, s=str(count), color='red', fontsize=12)
    
for num, label, val in zip([0,1], ['Female', 'Male'], gender_count.PassengerId):
    plt.text(x=num-0.05, y=val/2, s=str(label), color='white', fontsize=14 )

plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender count in Titanic', color='blue', fontsize=16)

plt.show()
```

# Data Visualization

```python
# titanic survived and not survived

plt.pie(titanic['Survived'].value_counts(), labels=['Dead', 'Survived'], autopct='%.2f')
plt.show()
```

```python
titanic['Age_group'] = titanic['Age'].map(lambda x: 'kids' if x<18 else 'adult' if x>=18 else 'unknown')
```

```python
sns.catplot('Sex', col='Age_group', kind='count', data=titanic, hue='Survived')
plt.show()
```

```python
# passenges survived per embarked

sns.catplot('Embarked', col='Survived', kind='count', data=titanic)
plt.show()
```

```python
# survival ratio of cabin vs non cabin


titanic['Cabin'].fillna('No_Cabin', inplace=True)

sns.catplot('Survived', col='Cabin', kind='count', data=titanic, sharey=False, aspect=0.7, col_wrap=5)
plt.show()
```

```python
# correlation between ticket and fare


titanic['fare_group']=pd.qcut(titanic['Fare'],10,range(1,11))
survivalRatio=titanic.groupby('fare_group')['Survived'].mean().reset_index()
plt.scatter(survivalRatio['fare_group'],survivalRatio['Survived'])
plt.show()
```
