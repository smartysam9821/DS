```python
def is_adult(age):
   if age >= 18:
       return 'adult'
   else:
       return 'kid'
```

```python
titanic.dropna(subset=['Age'], inplace=True)
```

```python
titanic['is_adult'] = titanic['Age'].map(lambda x: is_adult(x))
```
